```python
lat = 45.424721 
long = -75.695000
ottawa_map = folium.Map(location=[lat, long], zoom_start=12)

# display the map of Ottawa
ottawa_map
```

```python
ward_layer  = 'https://open.ottawa.ca/datasets/ottawa::wards-2022-2026.geojson'
```

```python
folium.GeoJson(ward_layer).add_to(ottawa_map)
```

```python
ottawa_map
```

```python
folium.GeoJson(
    ward_layer,
    name='Wards of Ottawa',
    style_function=lambda feature: {
        'fillColor': '#228B22',  # Green color
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.25
    }
).add_to(ottawa_map)
```

```python
ottawa_map
```

```python
df[['Level_of_Injury','Councillor', 'Neighbourhood','y', 'x']]
```

```python
# instantiate a feature group for the incidents in the dataframe
incident = folium.map.FeatureGroup()

# loop through the shootings and add each to the incidents feature group
for lat, lng, label, label1,label2, label3  in zip(df.y, df.x, df['Neighbourhood'], \
                                           df['Councillor'], df['Occurred_Year'], df['Level_of_Injury']):
    
    
    # Format the popup content with HTML
    popup_content = f"""
    <div>
        <strong>Neighbourhood:</strong> {label}<br>
        <strong>Councillor:</strong> {label1}<br>
        <strong>Occurred Year:</strong> {label2}<br>
        <strong>Level of Injury':</strong> {label3}<br>
    </div>
    """
    

    incident.add_child(
        folium.vector_layers.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            popup = folium.Popup(popup_content, max_width=300),
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
ottawa_map.add_child(incident)
```
